```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("data.csv")

# Goals per season
plt.figure(figsize=(12,6))
sns.countplot(y='Season', data=df, order=df['Season'].value_counts().index)
plt.title("Goals per Season")
plt.xlabel("Count")
plt.ylabel("Season")
plt.show()

# Goals per competition
plt.figure(figsize=(12,6))
col=df['Competition'].value_counts()
plt.pie(col)
plt.title('Goals by competition')
plt.ylabel('')
sns.countplot(y='Competition', data=df, order=df['Competition'].value_counts().index)
plt.title("Goals per Competition")
plt.xlabel("Count")
plt.ylabel("Competition")
plt.show()

# Goals by playing position
plt.figure(figsize=(8,6))
sns.countplot(y='Playing_Position', data=df, order=df['Playing_Position'].value_counts().index)
plt.title("Goals by Playing Position")
plt.xlabel("Count")
plt.ylabel("Position")
plt.show()

# Goal types
plt.figure(figsize=(8,6))
sns.countplot(y='Type', data=df, order=df['Type'].value_counts().index)
plt.title("Goal Types")
plt.xlabel("Count")
plt.ylabel("Type")
plt.show()

# Top Assist Providers
plt.figure(figsize=(10,6))
assist_counts = df['Goal_assist'].value_counts().head(10)
sns.barplot(x=assist_counts.values, y=assist_counts.index)
plt.title("Top 10 Assist Providers")
plt.xlabel("Assists")
plt.ylabel("Player")
plt.show()
```
